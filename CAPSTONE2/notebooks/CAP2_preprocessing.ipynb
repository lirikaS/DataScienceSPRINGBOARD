```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Load Dataset
url = "https://archive.ics.uci.edu/static/public/697/data.csv"
df = pd.read_csv(url, sep=',')


df = pd.get_dummies(df, columns=['Course'], drop_first=True)
```

Magnitude Standardization (Scaling)
The rubric specifically asks about features ranging from 0–1 or 0–100. In your data, Age (up to 70) and Grades (0–20) have different scales. Using a scaler prevents the model from assuming Age is more "important" simply because the numbers are larger.

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# We only scale numeric features (exclude binary indicators and the target)
numeric_cols = ['Age at enrollment', 'Curricular units 1st sem (grade)', 'Curricular units 1st sem (approved)']

# Fitting and transforming
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

Train/Test Split
The Strategy: Because your data has a 32.1% Dropout rate (Class Imbalance), you must use the stratify parameter. This ensures the 32% ratio is maintained in both your training and testing sets, preventing a "lucky" split where the test set has no dropouts.

```python
from sklearn.model_selection import train_test_split

X = df.drop('Target', axis=1)
y = df['Target']

# Splitting 75% for training, 25% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
X_train.to_csv('X_train_preprocessed.csv', index=False)
X_test.to_csv('X_test_preprocessed.csv', index=False)
```

Conclusion:

Does my dataset have categorical data? 
Yes. Features like Gender and Scholarship are categorical but already binary-encoded. If I included 'Course', I applied pd.get_dummies to avoid the dummy variable trap.

Do my features have varying ranges? 
Yes. Age ranges up to 70 while grades are on a 0–20 scale. I applied StandardScaler so that the XGBoost/Random Forest models treat the variance in grades with the same mathematical weight as the variance in age.

Is my response variable continuous or categorical?
The response variable (Dropout vs. Retained) is Categorical. This means I am performing a Classification task, and my success will be reasured by Recall and F1-Score rather than RMSE.
